# Online shopers intention
El conjunto de datos consta de vectores de características que pertenecen a 12.330 sesiones. 
Estos conjuntos de datos se formaron de modo que cada sesión pertenecería a un usuario diferente en un período de 1 año para evitar cualquier tendencia a una campaña específica, día especial, usuario perfil o período.

"Administrative", "Administrative Duration", "Informational", "Informational Duration", "Product Related" y "Product Related Duration" representan el número de diferentes tipos de páginas visitadas por el visitante en esa sesión y el tiempo total dedicado a cada una de estas categorías de páginas. Los valores de estas características se derivan de la información de URL de las páginas visitadas por el usuario y se actualizan en tiempo real cuando un usuario realiza una acción, por ejemplo pasar de una página a otra.
El "Bounce Rate", "Exit Rate" y "Page Value" las representan las métricas medidas por "Google Analytics" para cada página en el sitio de comercio electrónico. El valor de la función "Bounce Rate" para una página web se refiere al porcentaje de visitantes que ingresan al sitio desde esa página y luego lo abandonan ("Bounce") sin activar ninguna otra solicitud al servidor de análisis durante esa sesión.
El valor de la característica "Exit Rate" para una página web específica se calcula como para todas las visitas a la página, el porcentaje que fueron las últimas en la sesión.
La característica "Page Value" representa el valor promedio de una página web que un usuario visitó antes de completar una transacción de comercio electrónico.
La función "Special Day" indica la proximidad de tiempo de la visita del sitio a un día especial específico (por ejemplo, el Día de la Madre, el Día de San Valentín) en el que es más probable que las sesiones finalicen con una transacción. El valor de este atributo se determina considerando la dinámica del comercio electrónico, como la duración entre "order date" y "delivery date" (la fecha del pedido y la fecha de entrega). Por ejemplo, para el día de San Valentin, este valor toma un valor distinto de cero entre el 2 y el 12 de febrero, cero antes y después de esta fecha a menos que esté cerca de otro día especial, y su valor máximo de 1 el 8 de febrero.

El conjunto de datos también incluye "operating system", "browser", "region", "traffic type", "visitor type" como "returning" o "new visitor", un valor booleano que indica si la fecha de la visita es "weekend" (fin de semana) y "month of the yea"(el mes del año).

### El objetivo del analisis en este caso podria ser predecir las conductas de los visitantes dependiendo de variables como "SpecialDay", "region", "weekend" y con ello poder tomar decisiones que aumenten las ventas (por ejemplo pagando publicidades en google para que los visitantes de Android, suponiendo que descubrimso que son de donde más compran, tengan nuestra publicidad en sus feeds y busquedas del navegador).

## en esta oportunidad quie nsubió el df nos proporcionó cierta información que nos será de utilidad por eso la comparto. Es la explicación de los valores/columnas

1. gender: 0 (unknown), 1 (male), 2 (female)
2. C_api: gender extracted from WikiMedia API, codes as female / male / unknown
3. C_man: gender extracted from content coding, coded as 1 (male) / 2 (female) / 3 (unknown)
4. E_NEds: I index of stratum IJ (0,1,2,3)
5. E_Bpag: J index of stratum IJ (0,1,2,3)
6. firstDay: first edition in the Spanish Wikipedia (YYYYMMDDHHMMSS)
7. lastDay: last edition in the Spanish Wikipedia (YYYYMMDDHHMMSS)
8. NEds: total number of editions
9. NDays: number of days (lastDay-firstDay+1)
10. NActDays: number of days with editions
11. NPages: number of different pages edited
12. NPcreated: number of pages created
13. pagesWomen: number of edits in pages related to women
14. wikiprojWomen: number of edits in WikiProjects related to women
15. ns_user: number of edits in namespace user
16. ns_wikipedia: number of edits in namespace wikipedia
17. ns_talk: number of edits in namespace talk
18. ns_userTalk: number of edits in namespace user talk
19. s_content: number of edits in content pages
20. weightIJ: correcting weight for stratum IJ
21. NIJ: number of elements in stratum IJ
